<h1> Shor Algorithm via Quantum Inspire QASM</h1>


Seong Cho

This is implementation of Shor Algorithm via QASM at Quantum Inspire.


The code is as follows


    version 1.0

    qubits 5

    # start writing your code here

    #Hadamard gate on 0, 1, 2
    H q[0,1,2]

    #entangle 2 and 3
    CNOT q[2], q[3]

    #entangle 2 and 4
    CNOT q[2], q[4]

    #Hadamard gate on 1
    H q[1]

    # controlled rotation on 0 and 1 by 1/2 phi
    CR q[0], q[1], 1.57079632679

    # Hadamard on 0
    H q[0]

    # controlled rotation on 1 and 2 by 1/4 phi
    CR q[1], q[2], 0.78539816339

    # controlled rotation on 0 and 2 by 1/2 phi

    CR q[0], q[2], 1.57079632679

    measure q[0:1,2]
